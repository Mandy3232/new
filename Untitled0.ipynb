Check If A Number Is Even Or Odd

```
def check_num(n):
  if n%2==0:
    print("number is even")
  else:
    print("number is odd")

val=int(input("input a number"))
check_num(val)
```

Check If Number Is Prime

```
def check_prime(n):
  i=1
  j=0
  
  while(i<=n):
    if n%i==0:
       j+=1
    i+=1
    
  if j==2:
      print("number is prime")
  else:
      print("number is not prime")
      

val=int(input("input a number"))
check_prime(val)
```

Print Fibonacci Series

```
def fib_ser(n):
  a=0
  b=1
  print(a)
  print(b)
  for i in range(2,n):
    c=a+b
    a=b
    b=c
    print(c)
    
val=int(input("input a number"))
fib_ser(val)
```

Sum Of Squares Of Numbers From 1 To n

```
def sos(n):
  s=0
  for i in range (1,n+1):
    k=i**2
    s=s+k
  print(s)
val=int(input("input a number"))
sos(val)
```

Check If A Number Is Palindrome

```
def palin(n):
  new=0
  n1=n
  while(n!=0):
    r=n%10;
    new=new*10+r
    n=n/10
    print(n)
  
  if new==n1:
    print("number is palindrome")
  else:
    print("number is not palindrome")

val=int(input("input a number"))
palin(val)
```

Palindrome

```
def palindrome(num):
  if(num== "".join([x for x in num][::-1])):
    return True
  else:
    return False

  
value=input("enter a value")
palindrome(value)
```
